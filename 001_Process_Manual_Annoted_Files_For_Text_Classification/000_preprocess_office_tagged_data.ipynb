```python
# Function to read and create master dataframe: 
def read_tagged_files(path): 
     """
    Read and create a master dataframe from tagged files.

    This function reads the tagged files from the specified path and creates a master dataframe by concatenating
    the data from all the files.

    Args:
        path (str): The path to the directory containing the tagged files.

    Returns:
        pd.DataFrame: The master dataframe containing the combined data from all the tagged files.

    """
    tagged_files = os.listdir(path)
    master_df= pd.DataFrame()
    for file in tagged_files: 
        print(file)
        if file != '.ipynb_checkpoints':
            df = pd.read_excel(os.path.join(path, file))
            df = df[['review_id','full_review',
           'sentence','rating_managment_explicit',
           'disagreement_with_ratings', 'wrong_buying', 'zero_star']]
            master_df = pd.concat([master_df,df], ignore_index=True)
    return master_df 

def one_hot_encode(cnt):
     """
    Perform one-hot encoding.

    This function performs one-hot encoding based on the count value.

    Args:
        cnt (int): The count value.

    Returns:
        int: The one-hot encoded value (1 if count is greater than or equal to 1, 0 otherwise).

    """
    if cnt>=1: 
        return 1
    else: 
        return 0 
# for each label create dataset
def create_label_dataset(df, label): 
     """
    Create a dataset for a specific label.

    This function creates a dataset for a specific label by aggregating the data based on review ID and merging
    it with the review ID and full review columns from the original dataframe.

    Args:
        df (pd.DataFrame): The original dataframe.
        label (str): The label to create the dataset for.

    Returns:
        pd.DataFrame: The dataset for the specified label.

    """
    # seperate review_id, full_review: 
    # take only review id and full review
    df_1 = df[['review_id', 'full_review']] 
    df_1 = df_1.drop_duplicates()
    df_2 = df.groupby('review_id')[label].sum().reset_index()
    label_df = df_1.merge(df_2, on = 'review_id', how = 'inner')
    label_df[f'{label}_ohe']= label_df[label].apply(lambda x:one_hot_encode(x))
    return label_df
```

```python
# Function to read and create master dataframe: 
# for read reviews category, cannot use office_0 because label
# was developed later in the study
def read_tagged_files(path):
     """
    Read and create a master dataframe from tagged files.

    This function reads the tagged files from the specified path, removes the file 'amazon_reviews_office_0_all.xlsx',
    and creates a master dataframe by concatenating the data from the remaining files.

    Args:
        path (str): The path to the directory containing the tagged files.

    Returns:
        pd.DataFrame: The master dataframe containing the combined data from all the tagged files.

    """
    tagged_files = os.listdir(path)
    tagged_files.remove('amazon_reviews_office_0_all.xlsx')
    print(tagged_files)
    master_df= pd.DataFrame()
    for file in tagged_files: 
        print(file)
        if file != '.ipynb_checkpoints':
            df = pd.read_excel(os.path.join(path, file))
            print(df.columns)
            df = df[['review_id','full_review',
           'sentence','rating_managment_explicit',
           'disagreement_with_ratings', 'wrong_buying', 'zero_star',
           'wish_I_read_reviews ' ]]
            master_df = pd.concat([master_df,df], ignore_index=True)
    return master_df 

def one_hot_encode(cnt): 
    """
    Perform one-hot encoding.

    This function performs one-hot encoding based on the count value.

    Args:
        cnt (int): The count value.

    Returns:
        int: The one-hot encoded value (1 if count is greater than or equal to 1, 0 otherwise).

    """
    if cnt>=1: 
        return 1
    else: 
        return 0 
# for each label create dataset
def create_label_dataset(df, label): 
     """
    Create a dataset for a specific label.

    This function creates a dataset for a specific label by aggregating the data based on review ID and merging
    it with the review ID and full review columns from the original dataframe.

    Args:
        df (pd.DataFrame): The original dataframe.
        label (str): The label to create the dataset for.

    Returns:
        pd.DataFrame: The dataset for the specified label.

    """
    # seperate review_id, full_review: 
    # take only review id and full review
    df_1 = df[['review_id', 'full_review']] 
    df_1 = df_1.drop_duplicates()
    df_2 = df.groupby('review_id')[label].sum().reset_index()
    label_df = df_1.merge(df_2, on = 'review_id', how = 'inner')
    label_df[f'{label}_ohe']= label_df[label].apply(lambda x:one_hot_encode(x))
    return label_df
```
